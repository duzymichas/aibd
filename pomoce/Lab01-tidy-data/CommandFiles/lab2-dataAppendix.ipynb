```python
# Frequency table
pew = pd.read_csv(f"{DIR}pew.csv")
columns = pew.income
pd.crosstab(index=pew.reltrad, columns=columns)
```

```python
import matplotlib.pyplot as plt

# The whole bar chart
pew = pd.read_csv(f"{DIR}pew.csv")
pew = pew.value_counts(['reltrad', 'income'], sort=False)
pew.plot.bar(figsize=(50,50))
plt.show()
```

```python
# Bar charts from splited dataframe by religion
for x in range(0, len(pew), 10):
    pew[x:x+10].plot.bar(figsize=(10,10))
    plt.show()
```
